```python
import pandas as pd
import numpy as np
import scipy 
from scipy.signal import argrelextrema
from tqdm import tqdm, tnrange

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn.preprocessing import StandardScaler, MinMaxScaler

import os
import gc
import pickle
import warnings

warnings.filterwarnings('ignore')

plt.style.use('ggplot')

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = "Malgun Gothic"

os.chdir("../../")
root_path = os.getcwd()

raw_file_path = os.path.join(root_path, "Bigcon2020_raw_file")
csv_file_path = os.path.join(root_path, "BigCon_2020/csv_files")
pickle_file_path = os.path.join(root_path, "BigCon_2020/pickle_files")
```

```python
resales = pd.read_csv(os.path.join(csv_file_path, 'used_search_ratio.csv'))
```

```python
resales['period'] = pd.to_datetime(resales['period'])
```

```python
r1=(resales.loc[(resales.category=='중고') & (resales.period_cat==1)].ratio.values[-1]/resales.loc[(resales.category=='중고') & (resales.period_cat==2)].ratio.values[0])
r2=(resales.loc[(resales.category=='당근 마켓') & (resales.period_cat==1)].ratio.values[-1]/resales.loc[(resales.category=='당근 마켓') & (resales.period_cat==2)].ratio.values[0])
r3=(resales.loc[(resales.category=='중고나라') & (resales.period_cat==1)].ratio.values[-1]/resales.loc[(resales.category=='중고나라') & (resales.period_cat==2)].ratio.values[0])
r4=(resales.loc[(resales.category=='번개장터') & (resales.period_cat==1)].ratio.values[-1]/resales.loc[(resales.category=='번개장터') & (resales.period_cat==2)].ratio.values[0])
```

```python
resales.loc[(resales.category=='중고') & (resales.period_cat==2),'ratio']=resales.loc[(resales.category=='중고') & (resales.period_cat==2),'ratio']*r1
resales.loc[(resales.category=='당근 마켓') & (resales.period_cat==2),'ratio']=resales.loc[(resales.category=='당근 마켓') & (resales.period_cat==2),'ratio']*r2
resales.loc[(resales.category=='중고나라') & (resales.period_cat==2),'ratio']=resales.loc[(resales.category=='중고나라') & (resales.period_cat==2),'ratio']*r3
resales.loc[(resales.category=='번개장터') & (resales.period_cat==2),'ratio']=resales.loc[(resales.category=='번개장터') & (resales.period_cat==2),'ratio']*r4
```

```python
def replace_outlier(df, order=90):
    for category in tqdm(df.category.unique(), desc="1차 이상치 처리"):
        outlier_point = argrelextrema(df.loc[df.category==category,'ratio'].values, np.greater, order=order)
        outlier_index = df.loc[df.category==category].index[outlier_point]
        # method 1. 다음날 혹은 이전날 수치를 그대로 사용
        df.loc[df.index.isin(outlier_index),'ratio'] = np.nan
        df = df.fillna(method='bfill') # or 'ffill'
        
    for category in tqdm(df.category.unique(), desc="2차 이상치 처리"):
        outlier_point = argrelextrema(df.loc[df.category==category,'ratio'].values, np.greater, order=order)
        outlier_index = df.loc[df.category==category].index[outlier_point]
        # method 1. 다음날 혹은 이전날 수치를 그대로 사용
        df.loc[df.index.isin(outlier_index),'ratio'] = np.nan
        df = df.fillna(method='bfill') # or 'ffill'
        
    return df
```

```python
resales = replace_outlier(resales)
```
