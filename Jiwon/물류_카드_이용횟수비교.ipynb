```python
import pandas as pd
import numpy as np
```

```python
mul = pd.read_excel('2020 빅콘테스트_CJ올리브네트웍스_제공DB.xlsx')
card = pd.read_table('CARD_SPENDING_RESIDENT.txt')
```

```python
card = card.query('(GU_CD == 140) or (GU_CD == 350)').reset_index(drop=True)
mul = mul.query('CTPV_NM == "서울특별시"').reset_index(drop=True)
```

```python
card_f = card.query('(MCT_CAT_CD == 42) or (MCT_CAT_CD == 44)')
mul_f = mul.query('(DL_GD_LCLS_NM == "패션의류") or (DL_GD_LCLS_NM == "패션잡화")')
```

```python
mul_f['CTGG_CD'] = mul_f['CTGG_CD']*10
```

```python
dtdt = mul_f['DL_YMD']+20000000
mul_f['DL_YMD'] = dtdt.apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
card_f['STD_DD'] = card_f['STD_DD'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
card_f['YEAR_MONTH'] = card_f['STD_DD'].astype('str').str[:7]
mul_f['YEAR_MONTH'] = mul_f['DL_YMD'].astype('str').str[:7]
```

```python
card_f['WEEK_NUM'] = card_f['STD_DD'].dt.weekofyear
mul_f['WEEK_NUM'] = mul_f['DL_YMD'].dt.weekofyear
```

```python
card_f['YEAR'] = card_f['STD_DD'].dt.year
mul_f['YEAR'] = mul_f['DL_YMD'].dt.year
```

----------------

```python
df1 = card_f.groupby(['GU_CD','DONG_CD','YEAR','WEEK_NUM'])['USE_CNT'].mean().reset_index()
```

```python
df2 = mul_f.groupby(['CTGG_NM','CTGG_CD','HDNG_NM','HDNG_CD','YEAR','WEEK_NUM'])['INVC_CONT'].mean().reset_index()
```

```python
df1 = df1.rename(columns={'STD_DD':'DL_YMD','GU_CD':'CTGG_CD','DONG_CD':'HDNG_CD'})
```

```python
dli = df2[['CTGG_CD','CTGG_NM','HDNG_CD','HDNG_NM']].drop_duplicates()
```

```python
df1
```

```python
df0 = pd.merge(df1,df2,on=['CTGG_CD','HDNG_CD','YEAR','WEEK_NUM'])
df0.drop(columns=['CTGG_CD','HDNG_CD'],inplace=True)
```

```python
df0
```

```python
on_df2019 = pd.pivot_table(data=df0.query('YEAR == 2019'), index='HDNG_NM',columns='WEEK_NUM',values='INVC_CONT').fillna(0)
on_df2020 = pd.pivot_table(data=df0.query('YEAR == 2020'), index='HDNG_NM',columns='WEEK_NUM',values='INVC_CONT').fillna(0)
off_df2019 = pd.pivot_table(data=df0.query('YEAR == 2019'), index='HDNG_NM',columns='WEEK_NUM',values='USE_CNT').fillna(0)
off_df2020 = pd.pivot_table(data=df0.query('YEAR == 2020'), index='HDNG_NM',columns='WEEK_NUM',values='USE_CNT').fillna(0)
```

```python
onoff_2019 = on_df2019/off_df2019 #2019년의 온라인 이용률(WEEK)
onoff_2020 = on_df2020/off_df2020 #2020년의 온라인 이용률(WEEK)
```

```python
on_off = onoff_2020/onoff_2019
on_off = on_off.fillna(0)
# 1> : 코로나 이전에 비해 코로나 이후에 온라인에서 옷 쇼핑을 덜 함
# 1< : 코로나 이후에 온라인에서 옷쇼핑을 더 함
```

```python
on_off['MEAN'] = on_off.mean(axis=1)
```
